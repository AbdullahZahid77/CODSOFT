```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/IMDb Movies India.csv', encoding='latin-1')
```

```python
data.isnull().any()
```

```python
data.drop(columns=['Name'], inplace=True)
data
```

```python
data.dtypes
```

```python
null_count = data['Year'].isnull().sum()
null_count
```

```python
data.dropna(subset=['Year'], inplace=True)
```

```python
data
```

```python
# Remove the parentheses and convert to integers
data['Year'] = data['Year'].str.extract('(\d+)').astype(int)
```

```python
data.dtypes
data
```

```python
null_count = data['Duration'].isnull().sum()
null_count
```

```python
# Remove "mins" from the "duration" column
data['Duration'] = data['Duration'].str.replace(' min', '')

data.dtypes
```

```python
data['Duration'] = data['Duration'].astype(float).astype('Int64')
```

```python
data
```

```python
duration_mean=data['Duration'].mean()
duration_mean=duration_mean.astype(float).astype('int')
```

```python
data['Duration'].fillna(duration_mean, inplace=True)
```

```python

data
```

```python
rating_mean= data['Rating'].mean()
data['Rating'].fillna(rating_mean,inplace=True)
```

```python
data
```

```python
data = data[data['Votes'] != "$5.16M"]
data['Votes'] = data['Votes'].str.replace(',', '').astype(float)
```

```python
data
```

```python
votes_mean= data['Votes'].mean()
data['Votes'].fillna(votes_mean,inplace=True)
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_encoder = OrdinalEncoder()
data['Director'] = ordinal_encoder.fit_transform(data[['Director']])
```

```python
data
```

```python
data['Actor 1'] = ordinal_encoder.fit_transform(data[['Actor 1']])
data['Actor 2'] = ordinal_encoder.fit_transform(data[['Actor 2']])
data['Actor 3'] = ordinal_encoder.fit_transform(data[['Actor 3']])
```

```python
data
```

```python
actor1_mean= data['Actor 1'].mean()
data['Actor 1'].fillna(votes_mean,inplace=True)

actor2_mean= data['Actor 2'].mean()
data['Actor 2'].fillna(votes_mean,inplace=True)

actor3_mean= data['Actor 3'].mean()
data['Actor 3'].fillna(votes_mean,inplace=True)
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
data['Genre'] = ordinal_encoder.fit_transform(data[['Genre']])

genre_mean=data['Genre'].mean
data['Genre'].fillna(genre_mean,inplace=True)

data['Director'].fillna(0,inplace=True)
```

```python
data
```

```python
data.dtypes
```

```python
y=data['Rating']
X=data.drop('Rating',axis=1)
```

```python
X
```

```python
X=data.drop('Genre',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# X_train: Training features
# X_test: Testing features
# y_train: Training labels
# y_test: Testing labels
```

```python
X_train.isnull().any()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)

X_train.isnull().any()
```

```python
X_train
```

```python
rf_regressor.fit(X_train, y_train)
y_pred = rf_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse

#the mean absolute error is very small, which proves that the model is learning perfectly
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
```
